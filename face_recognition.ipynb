```python
import numpy as np
import cv2
```

```python
cascade_face = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_alt2.xml')
```

```python
cap = cv2.VideoCapture(0)
```

```python
while(True):
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = cascade_face.detectMultiScale(gray, scaleFactor = 1.5, minNeighbors = 5)
    for (x,y,w,h)in faces:
        print(x,y,w,h)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]
        img_item = "my-img.png"
        cv2.imwrite(img_item, roi_gray)
        color = (255,0,0)
        st = 2
        end_x = x+w
        end_y = y+h
        cv2.rectangle(frame, (x,y), (end_x, end_y),color,st)
    
    cv2.imshow('frame',frame)
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
    
```
